# Visualization of Diamonds data set

```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the file and create a dataframe
dataframe = pd.read_csv(r'C:\Users\Arakala Lahari\Desktop\DS project\diamonds.csv')
```

```python
#print the shape of the dataframe
print(dataframe.shape)
```

```python
#printing the dtypes of each column
print(dataframe.dtypes)
```

```python
#describe the dataframe
print(dataframe.describe())
```

```python
#Finding any null values in the data
dataframe.isna().sum()
```

```python
#printing the first 10 rows of the dataframe
print(dataframe.head(10))
print('Done')
```

```python
#Calculate the sum price of all the diamonds
sum = dataframe.price.sum()
print(f"The total Value of Diamonds is : {sum:,}$")
```

```python
#Calculate the mean price of all diamonds
mean = dataframe.price.mean()
print(f"The mean Value of Diamonds is : {mean:,.2f}$")
```

```python
#Get extra info about carat in diamonds
describe = dataframe.carat.describe()
print(describe)
```

```python
#Now let's visualize the data
# Let's see the clarity compared to the size,
# IF , VVS1, VVS2, VS1, VS2,SI1, I1 (from the purest to the least pure)

carat = dataframe.carat
clarity = dataframe.clarity

plt.scatter(clarity, carat)
plt.title('Diamond clarity compared to size in carat')
plt.xlabel('Clarity')
plt.ylabel('Carat')
plt.savefig("claritytocarat1.png")
```

```python
#Now let's visualize the data
# Let's see the carat compared to cut
carat = dataframe.carat
cut = dataframe.cut

plt.scatter(cut, carat)
plt.title('Diamond cut compared to size in carat')
plt.xlabel('Cut')
plt.ylabel('Carat')
plt.savefig("cuttocarat1.png")
```

```python
#We'll count the number of Diamonds for each Clarity Type
#We count the number for each type of clarity using value_counts index and values
clarityind = dataframe.clarity.value_counts().index.tolist()
claritycount = dataframe.clarity.value_counts().values
print(clarityind)
print(claritycount)
```

```python
#We visualize them with matplotlib
# IF , VVS1, VVS2, VS1, VS2,SI1, I1 (from the purest to the least pure)
plt.bar(clarityind, claritycount)
plt.title('Distribution of Diamonds according to their clarity')
plt.xlabel('Clarity')
plt.ylabel('Number of Diamonds')
plt.savefig("countClarity2.png")
```

```python
#Count the number of Diamonds for each color
colorind = dataframe.color.value_counts().index.tolist()
colorcount = dataframe.color.value_counts().values.tolist()
print(colorind)
print(colorcount)
```

```python
#Let's visualize this with a bar plot
# Note that D is the best color and J is the worst
plt.bar(colorind, colorcount)
plt.title('Distribution of Diamonds according to their clarity')
plt.xlabel('Color')
plt.ylabel('Number of Diamonds')
plt.savefig("countcolors3.png")
```
